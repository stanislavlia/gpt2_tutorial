```python
import torch
from transformers import GPT2LMHeadModel
import matplotlib.pyplot as plt

# Check if CUDA is available
if torch.cuda.is_available():
    print("GPU is available!")
    print("GPU Name:", torch.cuda.get_device_name(0))
else:
    print("GPU is not available.")
```

```python
model_hf = GPT2LMHeadModel.from_pretrained("gpt2") #124m 

state_dict = model_hf.state_dict()


for k,v in state_dict.items():

    print(k, v.shape)
```

```python
plt.imshow(state_dict["transformer.wpe.weight"], cmap="Greens")
```

```python

#Positional embeddings are learned from data in GPT2
plt.figure(figsize=(15, 5))
plt.plot(state_dict["transformer.wpe.weight"][:, 10])
plt.plot(state_dict["transformer.wpe.weight"][:, 50])
plt.plot(state_dict["transformer.wpe.weight"][:, 70])
plt.plot(state_dict["transformer.wpe.weight"][:, 200])
plt.plot(state_dict["transformer.wpe.weight"][:, 1])
```
